```python
import glob

import matplotlib.pyplot as plt

# Find all txt files ending with (accuracy).txt
accuracy_files = glob.glob("./*(accuracy).txt")


for file in accuracy_files:
    # Use filename (without extension and './') as label
    label = file.replace("./", "").replace("(accuracy).txt", "").strip(" _-")
    # Load data (assuming one value per line)
    with open(file, "r") as f:
        data = [float(line.strip()) for line in f if line.strip()]
    plt.plot(data, label=label)

plt.xlabel("Iteracin")
plt.ylabel("Rendimiento")
plt.axis([0, 50, 0.4, 1])
plt.legend()
plt.show()
```
